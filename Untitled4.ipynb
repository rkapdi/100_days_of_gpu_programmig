## What does a  photon look like?
### Light wave of 550nm (Visible Light)

```python
import math
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Create an array of values for the x position.
x = np.linspace(-5000,5000,10001)  # This creates values from -5000 to 5000 in steps of 1 and holds them in x  These values are in nm.
# Create another array to hold the values for the wave amplitude at each x.
y = np.linspace(0,0,10001) # Creates the range of the wave and the number of waves.
```

```python
# Calculate the amplitude of the wave assuming wavelength 550nm.
y = np.sin((2 * math.pi * x) / 550) # Converting wavelenght from nm to rads.
```

```python
# Create the plot
plt.plot(x, y)
# Add a label for the x-axis
plt.xlabel("Wavelength (nm)")
# Add a label for the y-axis
plt.ylabel("Amplitude (E = (m × 10⁻⁹)V)")
# Add a caption below the x-axis
plt.text(0.5, -0.2, "Fig1. A plot of photon amplitude versus wavlength", transform=plt.gca().transAxes, ha='center')
# Show the plot
plt.show()
```

```python
c = np.linspace(-10,10,21)    #create a range of wavelengths (Δλ=21 nm)
y = np.linspace(0,0,10001)
for i in c:
   k = 2*math.pi/(550+i)  # k is related to the momentum
   y= y + np.sin( k*x)   #  convert nm to radians and add the waves
# Create the plot
plt.plot(x, y)
# Add a label for the x-axis
plt.xlabel("Wavelength (nm)")
# Add a label for the y-axis
plt.ylabel("Amplitude (E = (m × 10⁻⁹)V)")
# Add a caption below the x-axis
plt.text(0.5, -0.2, "Fig2. A plot of photon amplitude versus wavlength of range (Δλ=21 nm).", transform=plt.gca().transAxes, ha='center')
# Show the plot
plt.show()
```

## Investigating The Relationship Between Wavelength & Momentum

The connection between wavelength and momentum is a bit like a trade-off.
If you try to be very sure about one (like knowing the exact momentum of a wave), you become very uncertain about the other (you don't know where the wave is).
It's a fundamental rule in the world of tiny things that adds a touch of mystery to the behavior of particles and waves.

## Prediction
**What will happen to the plot as (-10, 10, 21) is replaced with greater and greater range of wavelengths (i.e. more uncertainty in the momentum)?**

I think as we increase the range of wavelength (increasing uncertainity in momentum), the plot will change from a well defined wave packet to a chaotic pattern.

If we reduce the range, the pattern will be well defined, and conversely, it will get chaotic.
